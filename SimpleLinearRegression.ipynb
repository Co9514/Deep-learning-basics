```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
X = tf.placeholder(tf.float32, name = "X",shape = [None])
Y = tf.placeholder(tf.float32, name = "Y",shape = [None])

# 입력
x_input = [1,2,3,4]
# 출력
y_output = [2.3, 4.3, 6.3, 8.3]

W = tf.Variable(2.9)
b = tf.Variable(1.0)
```

```python
# 입력과 출력에 대한 가설을 수식으로 표현 (학습모델)
hypothesis = W * X + b

# 손실값을 측정하기 위한 손실함수
cost = tf.reduce_mean(tf.square(hypothesis - y_output))

# 0.01씩 하강하며 학습
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.001)

# 손실값이 최소가 되는 W 와 b (가중치와 편향)을 찾기 위해 역전파를 실행
train_op = optimizer.minimize(cost)
x_arr= []
y_arr=[]
z_arr=[]
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10000+1):        
        _,cost_val = sess.run([train_op,cost],feed_dict = {X:x_input,Y:y_output})
        if step % 10 == 0 :
            x_arr.append(sess.run(W))
            y_arr.append(cost_val)
            z_arr.append(sess.run(b))
            print(step, "손실값 : ", cost_val,"W : ",sess.run(W),"b : ",sess.run(b))
    print("\n=== Test ===")
    print("x: 5, Y : ", sess.run(hypothesis,feed_dict={X:[5]}))
    print("x: 2.5, Y : ", sess.run(hypothesis,feed_dict={X:[2.5]}))
```

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt 

mpl.rcParams['legend.fontsize'] = 10

fig = plt.figure()
ax = fig.gca(projection='3d')
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100) # 각도의 범위는 -4파이 에서 +4파이
ax.plot(x_arr, z_arr, y_arr, label='parametric curve')		# 위에서 정의한 x,y,z 가지고 그래프그린거다.
ax.legend()# 오른쪽 위에 나오는 글자 코드다. 이거 없애면 글자 사라진다. 없애도 좋다.

plt.show()
```
